## 1. Explore the data set and drop any rows with missing values.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

recent_grads = pd.read_csv('recent-grads.csv') #read csv
print(recent_grads.head())
```

```python
print(recent_grads.iloc[0]) #print first entry
```

```python
print(recent_grads.tail())
```

```python
recent_grads = recent_grads.dropna()
```

## 3. Create a for loop that plots the histogram of sample size, median, employed, full time, unemployment rate, men and women

```python

columns = ['Sample_size','Median','Employed','Full_time','Unemployment_rate','Men','Women']
fig = plt.figure(figsize = (10,20))
for i in range(1,8):
    ax = plt.subplot2grid((8,2),(i,0),rowspan=1,colspan=1)
    ax.hist(recent_grads[columns[i-1]])
fig.show()
```
